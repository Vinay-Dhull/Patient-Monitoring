```python
import pandas as pd
import os

# Directory containing the CSV files with complete data
input_directory = r'C:\Users\sathv\OneDrive\Desktop\Major Project\uqvitalsignsdata\case01\fulldata'  # Replace with your directory path
output_file = r'C:\Users\sathv\OneDrive\Desktop\Major Project\uqvitalsignsdata\combined_output.csv'  # Name of the output file

# List of columns to select (ensure 'Time' is the first column)
selected_columns = ['Time', 'HR', 'Pulse', 'SpO2', 'Perf', 'etCO2', 'imCO2', 'etN2O', 'inN2O', 
                    'MAC', 'etO2', 'inO2', 'Tidal Volume', 'Minute Volume', 'RR', 'ECG', 'Pleth', 
                    'CO2', 'AWP', 'AWF', 'AWV'] 

# List to hold data from each file
data_frames = []

# Loop through all files in the directory
for filename in os.listdir(input_directory):
    if filename.endswith('.csv'):
        file_path = os.path.join(input_directory, filename)
        # Read only the selected columns
        df = pd.read_csv(file_path, usecols=selected_columns)
        data_frames.append(df)

# Concatenate all data frames in the list
combined_df = pd.concat(data_frames)

# Reorder columns to ensure 'Time' is the first column
combined_df = combined_df[selected_columns]

# Save the combined data with selected columns to a new CSV file
combined_df.to_csv(output_file, index=False)

print(f"CSV with selected columns and 'Time' as the first column saved as {output_file}")
```
